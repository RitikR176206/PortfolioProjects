```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'Zomato data .csv')
df
```

```python
df.head()
```

```python
# Let’s convert the data type of the “rate” column to float and remove the denominator.
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)
df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
''' CONCLUSION: Dining restaurants primarily accept offline orders,
    whereas cafes primarily receive online orders. 
    This suggests that clients prefer to place orders in person at restaurants, 
    but prefer online ordering at cafes.
'''
```
